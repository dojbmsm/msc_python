```python
from numpy.ma.extras import average

scores = [78, 85, 62, 90, 55, 88, 73]
max(scores)
min(scores)
average(scores)
average = sum(scores) / len(scores)

print(average)

higher_scores = list(filter(lambda x: x > 75, scores))
print(higher_scores)
scores.sort(reverse=True)
print(scores)
scores_tuple = tuple(scores)
print(scores_tuple[0])
print(scores_tuple[-1])
print(len(scores_tuple))
print(len(list(filter(lambda x: x > average, scores))))
```
